```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from ggplot import *
pd.set_option('display.height', 10500)
pd.set_option('display.max_rows', 10500)
```

```python
years = range(1880, 2014)

dirName = '/Users/gabrielaliquo/pythongame/names/'

pieces = []
columns = ['name','sex','births']

for year in years:
    path = dirName + 'yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    frame['age'] = 2015 - year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
```

```python
print names.head(n=135)
```

```python
NameList = ['Matthew', 'Erica', 'Francis', 'Zachary', 'Emily', 'Gabriel', 'Audrey', 'Judith', 'Erin', 'James']
```

```python
Hoovers = names[names.name.isin(NameList)]
```

```python
agepivot = Hoovers.pivot_table('births', index='age', columns='name')
print agepivot.head(n=135)
```
