```python
import numpy as np
import pandas as pd
import pandas.io.data as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aapl = web.DataReader('AAPL', data_source='google', start= '01/01/2014', end = '03/26/2015')
```

```python
print aapl.head(n=5)
print aapl.tail(n=5)
```

```python
aapl[['Open', 'Close']].plot(subplots=True, color='r', figsize=(10,10))
```

```python
print aapl.head(n=10)
```

```python
aapl['change'] = ((aapl.Close - aapl.Open)/aapl.Open)*100
aapl['name'] = 'Apple'
print aapl.head(n=10)
```

```python
#aaplnew =aapl.pivot_table('Volume', index='name', columns='Date')
#aapl_pivot.plot(title = 'Apple Percent Change Daily')
new_aapl = aapl
new_aapl.change.dropna().plot()
```

```python
print new_aapl.head(n=10)
```

```python
aapl['Daily_diff'] = 100*(aapl.Close - aapl.Close.shift(1))/(aapl.Close.shift(1))
aapl['Overnight_chng'] = 100*(aapl.Open.shift(-1) - aapl.Close)/(aapl.Close)
print aapl.head(n=10)
print aapl.tail(n=10)
```

```python
aapl.boxplot('Overnight_chng', return_type='axes')
```
